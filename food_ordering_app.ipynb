```python
app_name = "Patty's"
```

```python
menu = {'sku1':{'name': 'Hamburger', 'price': 6.51}, 
        'sku2':{'name': 'Cheeseburger', 'price': 7.75}, 
        'sku3':{'name': 'Milkshake', 'price': 5.99}, 
        'sku4':{'name': 'Fries', 'price': 2.39},
        'sku5':{'name': 'Sub', 'price': 5.87}, 
        'sku6':{'name': 'Ice Cream', 'price': 1.55}, 
        'sku7':{'name': 'Fountain Drink', 'price': 3.45}, 
        'sku8':{'name': 'Cookie', 'price': 3.15}, 
        'sku9':{'name': 'Brownie', 'price': 2.46}, 
        'sku10':{'name': 'Sauce', 'price': 0.75}}
```

```python
# Dictionary test

print(menu['sku1'])
print(menu['sku2']['name'])
print(menu['sku3']['price'])
```

```python
actions = {1: "Add a new menu item to cart",
           2: "Remove an item from the cart",
           3: "Modify a cart item's quantity",
           4: "View cart",
           5: "Checkout",
           6: "Exit"}           
```

```python
sales_tax = 0.03
```

```python
cart = {}
```

```python
# Function to display the menu

def display_menu():
    print("\nHere is the menu:")
    print("")
    for item in menu:
        item_number = item[-1]
        item_name = menu[item]['name']
        item_price = (menu[item]['price'])
        print(f"({item_number}) {item_name}: ${item_price}")
```

```python
display_menu()
```

```python
# A function to add the entered SKU number and quantity to the cart.

def add_to_cart(sku, quantity):
    if sku in menu:
        if sku in cart:
            cart[sku] += quantity
        else:
            cart[sku] = quantity
        print("\nAdded ", quantity, 
              " of", menu[sku]['name'], 
              " to the cart.")
    else:
        print("\nSorry! That SKU is not valid.")
```

```python
# Testing the add_to_cart function

add_to_cart("sku1", 2)  # Should add 2 Hamburgers
add_to_cart("sku2", 1)  # Should add 1 Cheeseburger
add_to_cart("sku11", 1)  # Should print error message (invalid SKU)

print(cart)  # See what's in cart now
```

```python
# A function to remove the entered SKU item from the cart if it exists.

def remove_from_cart(sku):
    if sku in cart:
        removed_item = cart.pop(sku)
        print(f"\nRemoved {menu[sku]['name']} from the cart.")
    else:
        print("\nSorry! That SKU is not valid.")
```

```python
# Testing the remove_from_cart function

remove_from_cart("sku1")

print(cart)
```

```python
# A function to modify the cart by updating the entered SKU item with the entered quantity.

def modify_cart(sku, quantity):
    if sku in cart:
        if quantity > 0:
            cart[sku] = quantity
        else:
            remove_from_cart(sku)
        print(f"\nModified {menu[sku]['name']} quantity to {quantity} in the cart")
    else:
         print("\nSorry! That SKU is not valid.")
```

```python
# Testing the modify_cart function

modify_cart("sku2", 5)

print(cart)

modify_cart("sku9", 0)

print(cart)
```

```python
# A function to:
# 1. Display the current cart items and quantities.
# 2. Calculate and display the current subtotal of items in the cart.

def view_cart():
    print("\n*** Cart Contents ***")
    subtotal = 0
    for sku in cart:
        if sku in menu:
            quantity = cart[sku]
            cost = quantity * (menu[sku]['price'])
            subtotal += cost
            print(f"\n- {quantity} x {menu[sku]['name']}: ${cost}")
    taxes = subtotal * sales_tax
    print(f"\nSubtotal: ${subtotal:.2f}")
    print(f"\nTaxes: ${taxes:.2f}")
    print(f"\nTOTAL: ${taxes + subtotal:.2f}")
```

```python
# Testing the view_cart function

remove_from_cart('sku9')
add_to_cart('sku9', 12)

view_cart()
```

```python
# A function for the customer to check out

def checkout():
    print("\n+++ CHECKOUT +++")
    view_cart()
    print("\nYour order has been received!")
```

```python
# Testing the checkout function

checkout()
```

```python
# A function to query the customer for order info
# Gets reused for multiple types of app actions (see other functions)

def get_sku_and_quantity(sku_prompt, quantity_prompt = None):
    sku_number = input(sku_prompt)
    item_sku = f"sku{sku_number}"
    if quantity_prompt:
        item_quantity = input(quantity_prompt)
        if item_quantity.isdigit():
            item_quantity = int(item_quantity)
        else:
            item_quantity = 1
        return(item_sku, item_quantity)
    return(item_sku)
```

```python
# A function for running the app loop

def ordering_loop():
    print(f"Welcome to {app_name}!\n")
    is_running = True
    while is_running:
        print("\n*** Ordering Actions ***\n")
        for key, action in actions.items():
            print(f"({key}) {action}")
        user_prompt = input("\nEnter the number of the action you'd like to take: ")
        if user_prompt == '1':
            display_menu()
            print('\n')
            sku, quantity = get_sku_and_quantity('Enter sku:', 'Enter quantity:')
            add_to_cart(sku, quantity)
        elif user_prompt == '2':
            display_menu()
            print('\n')
            sku = get_sku_and_quantity('Enter sku:')
            remove_from_cart(sku)
        elif user_prompt == '3':
            display_menu()
            print('\n')
            sku, quantity = get_sku_and_quantity('Enter sku:', 'Enter quantity:')
            modify_cart(sku, quantity)
        elif user_prompt == '4':
            view_cart()
        elif user_prompt == '5':
            checkout()
            is_running = False
        elif user_prompt == '6':
            is_running = False
        else:
            print("\nPlease pick an action number from the list (1-5)")
```

```python
# Start the app!

ordering_loop()
```
